```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

img_size = 512
stride = 8
base_size = 32
half = base_size / 2

fig, ax = plt.subplots(1, figsize=(6, 6))
ax.set_xlim(0, img_size)
ax.set_ylim(0, img_size)
ax.invert_yaxis()

# Draw a few anchors at different positions
for i, j in [(10, 10), (20, 20), (30, 30)]:
    cx = (j + 0.5) * stride
    cy = (i + 0.5) * stride
    anchor = patches.Rectangle((cx - half, cy - half), base_size, base_size,
                               linewidth=2, edgecolor='red', facecolor='none')
    ax.add_patch(anchor)
    ax.plot(cx, cy, 'bo')  # center

plt.title("Anchors at Feature Map Locations (P3)")
plt.grid(True)
plt.show()
```
